## 6*. Дизамбигуация в Mystem (2 балла)

*необязательное задание на 10 баллов

Одна из фичей майстема - контекстная дизамбигуация (т.е. определение правильного разбора в зависимости от контекста).

Запустите mystem на очищенном тексте с дизамбигуацией и без и проверьте есть ли какие-то различия.

````python
import json

with open('text_nocontext.jsonlines', 'r', encoding='utf-8') as f:
    mystem_nocont_words = [json.loads(line) for line in f]
````

````python
with open('text_withcontext.jsonlines', 'r', encoding='utf-8') as f:
    mystem_withcont_words = [json.loads(line) for line in f]
````

````python
from collections import namedtuple

Analysis = namedtuple('Analysis', ['text', 'lex', 'gr'])

def collect_mystem_analyses(mystem_words, use_first_n=None):
    shorter_mystem_words = []
    
    for par in mystem_words:
        for tok in par:
            text = tok['text']
            if not text.strip():
                continue

            lex = [text]
            gr = ['']

            if 'analysis' in tok and len(tok['analysis']) > 0:
                lex = [option['lex'] for option in tok['analysis']][:use_first_n]
                gr = [option['gr'] for option in tok['analysis']][:use_first_n]

            shorter_mystem_words.append(Analysis(text, lex, gr))
    
    return shorter_mystem_words
            
    # print(*[(an['text'], an.get('analysis') and (an['analysis'])
    #         for an in sent],
    #       sep='\n', end=f'\n{"-" * 100 + "PARAGRAPH" + "-" * 100}\n'
    # )
````

````python
Comparison = namedtuple('Comparison', ['text', 'len_1', 'len_2', 'are_lex_equal', 'are_gr_equal'])


def compare_mystem_analyses(analyses_1, analyses_2):
    result = []
    
    for an_1_, an_2_ in zip(analyses_1, analyses_2):
        comp = Comparison(an_1_.text, len(an_1_.lex), len(an_2_.lex), an_1_.lex==an_2_.lex, an_1_.gr==an_2_.gr)
        
        result.append(comp)
    
    return result    
````

Что можно заметить сразу, это выбор одного анализа для *о* как предлога при использовании контекста. Это и по словарю был самый вероятный вариант, но контекст позволяет установить это однозначно. Ниже мы сравним результаты подробнее.

````python
print(*(collect_mystem_analyses(mystem_nocont_words[:2])[:10]), sep='\n')
````

````python
print(*(collect_mystem_analyses(mystem_withcont_words[:2])[:10]), sep='\n')
````

Сравним это подробнее

````python
mystem_nocont_ = collect_mystem_analyses(mystem_nocont_words)
mystem_withcont_ = collect_mystem_analyses(mystem_withcont_words)

comparison_res = compare_mystem_analyses(mystem_nocont_, mystem_withcont_)
````

````python
comparison_res[:10]
````

````python
check_n = 50

for i, comp in enumerate(comparison_res[:check_n]):
    if not comp.are_lex_equal:
        print(f"{mystem_withcont_[i]}\n\t\t{mystem_nocont_[i]}")
````

````python
stats = dict(diff=dict(max_val=-1, max_at=-1, min_val=None), mult_decrease_max=-1, mult_decrease_max_at=-1)

for i, comp in enumerate(comparison_res):
    diff = abs(comp.len_1 - comp.len_2)
    
    if diff > stats['diff']['max_val']:
        stats['diff']['max_val'] = diff
        stats['diff']['max_at'] = i
        
    stats['diff']['min_val'] = diff if stats['diff']['min_val'] is None else min(stats['diff']['min_val'], diff)
    
    mult_decrease = comp.len_1 / comp.len_2
    if mult_decrease > stats['mult_decrease_max']:
        stats['mult_decrease_max'] = mult_decrease
        stats['mult_decrease_max_at'] = i
        
print(stats)
````
